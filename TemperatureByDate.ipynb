### Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
plt.rcParams['figure.figsize'] = (20.0, 10.0)
```

### Read input data

```python
path= 'datasets/TemperatureByDate.csv'
data = pd.read_csv(path)
data['date'] = pd.to_datetime(data['date'])
data.set_index('date', inplace=True)
data.head()
```

```python
data.index.dtype
```

### Temperature by date

```python
data['temperature'].plot(title = "Temperature")
```

### Average temperature per year

```python
average_per_year = data.resample('Y').mean()
average_per_year.plot(title = "Average temperature per year")
```

### Minimum temperature per year

```python
minimum_per_year = data.resample('Y').min()
minimum_per_year.plot(title = "Minimum temperature per year")
```

### Maximum temperature per year

```python
maximum_per_year = data.resample('Y').max()
maximum_per_year.plot(title = "Maximum temperature per year")
```
